```python
import pandas as pd
```

- S1101 --> Household Types
- S1701 --> Poverty
- DP05 --> Median age and race by place
- S2301 --> Unemployment rate
- B19126 --> Median income by place

```python
household_types = pd.read_csv('data/ACS_17_1YR_S1101_with_ann.csv', header=1)
poverty = pd.read_csv('data/ACS_17_1YR_S1701_with_ann.csv', header=1)
age_race = pd.read_csv('data/ACS_17_1YR_DP05_with_ann.csv', header=1)
unemployment = pd.read_csv('data/ACS_17_1YR_S2301_with_ann.csv', header=1)
income = pd.read_csv('data/ACS_17_1YR_B19126_with_ann.csv', header=1)
```

```python
#Get lengths of dataframes so we know how much we're losing
print("household_types: ", len(household_types))
print("poverty: ", len(poverty))
print("age_race: ", len(age_race))
print("unemployment: ", len(unemployment))
print("income: ", len(income))
```

```python
#Drop unnecessary columns
household_types = household_types[[
    'Id',
    'Geography',
    'Total; Estimate; Total households',
    'Female householder, no husband present, family household; Estimate; AGE OF OWN CHILDREN - Households with own children of the householder under 18 years',
    'Male householder, no wife present, family household; Estimate; AGE OF OWN CHILDREN - Households with own children of the householder under 18 years',
    'Total; Estimate; AGE OF OWN CHILDREN - Households with own children of the householder under 18 years'
]]
poverty = poverty[[
    'Id',
    'Percent below poverty level; Estimate; Population for whom poverty status is determined'
]]
age_race = age_race[[
    'Id',
    'Estimate; SEX AND AGE - Median age (years)',
    'Percent; RACE - One race - White',
    'Percent; RACE - One race - Black or African American',
    'Percent; RACE - One race - American Indian and Alaska Native',
    'Percent; RACE - One race - Asian',
    'Percent; HISPANIC OR LATINO AND RACE - Total population - Hispanic or Latino (of any race)'
]]
unemployment = unemployment[[
    'Id',
    'Unemployment rate; Estimate; Population 16 years and over'
]]
income = income[[
    'Id',
    'Estimate; Median family income in the past 12 months (in 2017 inflation-adjusted dollars) -- - Married-couple family -- - With own children of the householder under 18 years',
    'Estimate; Median family income in the past 12 months (in 2017 inflation-adjusted dollars) -- - Other family -- - Male householder, no wife present -- - With own children of the householder under 18 years',
    'Estimate; Median family income in the past 12 months (in 2017 inflation-adjusted dollars) -- - Other family -- - Female householder, no husband present -- - With own children of the householder under 18 years'
]]
```

```python
from functools import reduce
datasets = [household_types, age_race, poverty, income, unemployment]
merged = reduce((lambda x,y: x.merge(y, on='Id', how='inner')), datasets)
```

```python
merged
```

Great, we didn't lose a single row! Now let's lint out every place with <100,000 households and calculate our other statistics.

```python
merged = merged.loc[
    merged['Total; Estimate; Total households'] > 100000
]
merged.dropna(axis=0, how='any',inplace=True)
merged
```

```python
#Convert to floats
for col in list(merged):
    if not col == 'Id' and not col == 'Geography':
        merged[col] = merged[col].astype('float')

#Calculate percent single-parent households with children <18
merged['Percent single parent households with children'] = (merged[
    'Female householder, no husband present, family household; Estimate; AGE OF OWN CHILDREN - Households with own children of the householder under 18 years'
] + merged[
    'Male householder, no wife present, family household; Estimate; AGE OF OWN CHILDREN - Households with own children of the householder under 18 years'
]) / merged[
    'Total; Estimate; AGE OF OWN CHILDREN - Households with own children of the householder under 18 years'
] * 100

merged['Percent single mother households with children'] = (
    merged[
        'Female householder, no husband present, family household; Estimate; AGE OF OWN CHILDREN - Households with own children of the householder under 18 years'
    ] / merged[
        'Total; Estimate; AGE OF OWN CHILDREN - Households with own children of the householder under 18 years'
    ]
) * 100

merged['Percent single father households with children'] = (
    merged[
        'Male householder, no wife present, family household; Estimate; AGE OF OWN CHILDREN - Households with own children of the householder under 18 years'
    ] / merged[
        'Total; Estimate; AGE OF OWN CHILDREN - Households with own children of the householder under 18 years'
    ]
) * 100
```
